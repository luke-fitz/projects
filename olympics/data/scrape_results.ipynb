```python
from bs4 import BeautifulSoup
import pandas as pd
import requests


def get_all_event_urls(main_url):
    response = requests.get(main_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all <a> tags where the link text is "details"
    event_urls = ["https://en.wikipedia.org/" + a['href'] for a in soup.find_all('a', string="details")]
    return event_urls
    

def get_final_df(event_url):
    response = requests.get(event_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the <h3> tag with id="Final"
    h3_tag = soup.find('h3', id="Final")

    # Find the first table after this <h3> tag
    table = h3_tag.find_next('table')

    # Read the table into a pandas DataFrame
    # Assuming the table has <tr> for rows and <td> for columns
    data = []
    for row in table.find_all('tr'):
        cols = [ele.text.strip() for ele in row.find_all(['td', 'th'])]
        data.append(cols)

    df = pd.DataFrame(data[1:], columns=data[0])  # Assuming first row is the header
    
    return df

# Load your HTML content
main_url = 'https://en.wikipedia.org/wiki/Athletics_at_the_2024_Summer_Olympics'
event_urls = get_all_event_urls(main_url)
df = pd.DataFrame()
for event_url in event_urls:
    print(event_url)
    try:
        df_event = get_final_df(event_url)
        df_event['url'] = event_url.replace('https://en.wikipedia.org//wiki/Swimming_at_the_2024_Summer_Olympics_%E2%80%93_', '')
        df = pd.concat([df, df_event], axis = 0)
    except:
        continue
```
